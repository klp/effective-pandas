## 5.1 Loading the data

Not only can `read_csv()` method read csvs, but it can also pull in zip files!

```python
import pandas as pd
```

```python
url = 'https://github.com/mattharrison/datasets/raw/master/data/vehicles.csv.zip'
```

```python
df = pd.read_csv(url)
```

```python
city_mpg = df.city08
highway_mpg = df.highway08
```

```python
city_mpg
```

```python
highway_mpg
```
